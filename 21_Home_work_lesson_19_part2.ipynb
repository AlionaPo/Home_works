```python
import numpy as np
import pandas as pd
import seaborn as sns
penguins = sns.load_dataset("penguins")

sns.displot(data=penguins, x="body_mass_g", hue="species", col="species")
```

```python
sns.histplot(data=penguins, x="species", hue="species", shrink=.8, alpha=.8, legend=False)
```

```python
sns.kdeplot(
    data=penguins, x="body_mass_g", hue='sex',
    fill=True, common_norm=False, palette="crest",
    alpha=.5, linewidth=0,)
```
